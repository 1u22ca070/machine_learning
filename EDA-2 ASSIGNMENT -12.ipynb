#Standard scaling works well when the data is symmetrically distributed around the mean.
#it centers the data and ensures all features are on the same scale,

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import sklearn
print(sklearn.__version__)
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
```

```python
for column in categorical_columns:
    unique_values=df[column].nunique()
    
    if unique_values<5:
        encoder=OneHotEncoder(sparse_output=False, drop='first')
        one_hot_encoded = encoder.fit_transform(df[[column]])
        encoded_df = pd.DataFrame(
            one_hot_encoded, 
            columns=[f"{column}_{race}" for race in encoder.categories_[0][1:]]
        )
        df = pd.concat([df.drop(column, axis=1), encoded_df], axis=1)
    else:
         encoder = LabelEncoder()
         df[column] = encoder.fit_transform(df[column])
```

```python
processed_file_path='processed_file.csv'
df.to_csv(processed_file_path,index=False)
print(f"/nprocessed data saved to '{processed_file_path}'.")
```

#One-Hot Encoding
#Pros
1.No Assumptions of Ordinality:
Treats categories as independent, which is beneficial when there’s no inherent order between them

2.Widely Compatible:
Works well with algorithms that rely on distance metrics, like Logistic Regression, Support Vector Machines, and Neural Networks.


#Cons
1.High Dimensionality:
For variables with many categories, it can create a large number of columns, leading to a curse of dimensionality, increased computational costs, and potential overfitting.
Sparse Matrix:

Results in a sparse matrix with many zero entries, which can be inefficient in terms of memory and storage.

#Label Encoding
#Pros

1.Simplicity:
Easy to implement, as it converts categories to numeric values directly.

2.Memory Efficient:
Uses a single column instead of creating multiple columns, saving memory and computational resources.

#Cons

Ordinal Assumption:

Implies an order or relationship between categories, which can mislead the model if such relationships don’t exist (e.g., treating "Apple" < "Banana" < "Cherry").
Bias Introduction:

Models that interpret numeric values as meaningful (e.g., decision trees or linear models) might assign importance to category values based on their magnitude.
